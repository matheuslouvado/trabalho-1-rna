```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('heart_failure_clinical_records_dataset.csv')
```

```python
df.head()
```

```python
X = df.drop('DEATH_EVENT', axis=1)
y = df['DEATH_EVENT']
```

```python
X.shape
X.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
y_enc = le.fit_transform(y)
```

```python
y_enc.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_enc, test_size=0.33)
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([ 
    keras.layers.Dense(24, input_shape=(12,), activation='relu', name='oculta1'),
    keras.layers.Dense(8, activation='relu', name='oculta2'),
    keras.layers.Dense(4, activation='relu', name='oculta3'),
    keras.layers.Dense(1, activation='sigmoid', name='saida')
])
model.summary()
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.001), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train,y_train, epochs = 200)
```

```python
model.evaluate(X_test,y_test)
```
